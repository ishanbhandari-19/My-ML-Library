```
a = km.K_means()
```

```
from sklearn.datasets.samples_generator import make_blobs
X,y = make_blobs(n_samples = 500,centers = 5,cluster_std = 0.60, random_state=0)
plt.scatter(X[:,0],X[:,1], s=10)
```

```
y_pred = a.predict(X,5)
```

```
plt.scatter(X[:,0],X[:,1],c=list(y_pred),cmap = 'viridis')
```

```
from sklearn.cluster import KMeans
km = KMeans(n_clusters=5,init = 'random',n_init = 10,max_iter=300,tol = 1e-04,random_state = 0)
y_km = km.fit_predict(X)
```

```
plt.scatter(X[:,0],X[:,1],c=list(y_km),cmap = 'viridis')
```
